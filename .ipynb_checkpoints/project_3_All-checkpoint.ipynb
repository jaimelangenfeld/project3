# Aggregate each year total crime per capita

```python
aggregate_df = merged_df[['City', '2009_Population', '2009_Violent_crime', '2009_Property_crime', '2019_Population', '2019_Violent_crime', '2019_Property_crime']].copy()
aggregate_df.set_index('City', inplace=True)
aggregate_df.head()
```

```python
# Calculate totals for each column
total_2009_population = aggregate_df['2009_Population'].sum()
total_2009_violent_crime = aggregate_df['2009_Violent_crime'].sum()
total_2009_property_crime = aggregate_df['2009_Property_crime'].sum()
total_2019_population = aggregate_df['2019_Population'].sum()
total_2019_violent_crime = aggregate_df['2019_Violent_crime'].sum()
total_2019_property_crime = aggregate_df['2019_Property_crime'].sum()

# Print totals for each column
print("2009 Population:", total_2009_population)
print("Total 2009 Violent Crime:", total_2009_violent_crime)
print("Total 2009 Property Crime:", total_2009_property_crime)
print("2019 Population:", total_2019_population)
print("Total 2019 Violent Crime:", total_2019_violent_crime)
print("Total 2019 Property Crime:", total_2019_property_crime)
```

```python
# Calculate crime rates
violent_crime_rate_2009 = round((total_2009_violent_crime / total_2009_population) * 1000, 2)
property_crime_rate_2009 = round((total_2009_property_crime / total_2009_population) * 1000, 2)
total_crime_rate_2009 = round(violent_crime_rate_2009 + property_crime_rate_2009, 2)
violent_crime_rate_2019 = round((total_2019_violent_crime / total_2019_population) * 1000, 2)
property_crime_rate_2019 = round((total_2019_property_crime / total_2019_population) * 1000, 2)
total_crime_rate_2019 = round(violent_crime_rate_2019 + property_crime_rate_2019, 2)

# Print crime rates
print("2009 Violent Crime Rate:", violent_crime_rate_2009)
print("2009 Property Crime Rate:", property_crime_rate_2009)
print("2009 Total Crime Rate:", total_crime_rate_2009)
print("2019 Violent Crime Rate:", violent_crime_rate_2019)
print("2019 Property Crime Rate:", property_crime_rate_2019)
print("2019 Total Crime Rate:", total_crime_rate_2019)
```

```python
# Create a dictionary with the crime rates
summary_data = {
    'Crime Type': ['Violent Crime', 'Violent Crime', 'Property Crime', 'Property Crime', 'Total Crime', 'Total Crime'],
    'Year': ['2009', '2019', '2009', '2019', '2009', '2019'],
    'Crime Rate': [violent_crime_rate_2009, violent_crime_rate_2019, property_crime_rate_2009,
                   property_crime_rate_2019, total_crime_rate_2009, total_crime_rate_2019]
}

# Create the DataFrame
summary_df = pd.DataFrame(summary_data)

# Reorder the Crime Type column
crime_type_order = ['Violent Crime', 'Property Crime', 'Total Crime']
summary_df['Crime Type'] = pd.Categorical(summary_df['Crime Type'], categories=crime_type_order, ordered=True)

# Display the summary DataFrame
summary_df
```

```python
# Create pivot table
pivot_df = summary_df.pivot(index='Crime Type', columns='Year', values='Crime Rate')

# Create bar chart
pivot_df.plot(kind='bar', figsize=(12, 8), color=['teal', 'firebrick'])
plt.title('Crime Rate Comparison (2009 vs 2019)')
plt.ylabel('Crime Rate (per 1,000)')
plt.xlabel('')
plt.xticks(rotation=0)
plt.legend(title='Year', loc='upper left')

plt.show()
```

# Biggest changes of crime in cities

```python
crimes=['Violent_crime','Murder_and_nonnegligent_manslaughter','Rape','Robbery','Aggravated_assault','Property_crime','Burglary','Larceny_theft','Motor_vehicle_theft','Arson']

#Crime rate by crime 2009
for crime in crimes:
    merged_df[f'2009_{crime}_Rate']=merged_df[f'2009_{crime}']/merged_df['2009_Population']*1000

#Crime rate by crime 2019
for crime in crimes:
    merged_df[f'2019_{crime}_Rate']=merged_df[f'2019_{crime}']/merged_df['2019_Population']*1000

#Percent change
for crime in crimes:
    merged_df[f'PercentChange_{crime}']=((merged_df[f'2019_{crime}_Rate']-merged_df[f'2009_{crime}_Rate'])/merged_df[f'2009_{crime}_Rate'])*100

#When calculating percent change, some rows had to divide by zero, created values 'inf'. Change these to 0
replace=['PercentChange_Violent_crime',
         'PercentChange_Murder_and_nonnegligent_manslaughter',
         'PercentChange_Rape',
         'PercentChange_Robbery',
         'PercentChange_Aggravated_assault',
         'PercentChange_Property_crime',
         'PercentChange_Burglary',
         'PercentChange_Larceny_theft',
         'PercentChange_Motor_vehicle_theft',
         'PercentChange_Arson']
merged_df[replace]=merged_df[replace].replace(np.inf,0)

#Change columnt names to make chart neater
merged_df['Violent Crime']=merged_df['PercentChange_Violent_crime']
merged_df['Murder']=merged_df['PercentChange_Murder_and_nonnegligent_manslaughter']
merged_df['Rape']=merged_df['PercentChange_Rape']
merged_df['Robbery']=merged_df['PercentChange_Robbery']
merged_df['Aggravated Assault']=merged_df['PercentChange_Aggravated_assault']
merged_df['Property Crime']=merged_df['PercentChange_Property_crime']
merged_df['Burglary']=merged_df['PercentChange_Burglary']
merged_df['Larceny-Theft']=merged_df['PercentChange_Larceny_theft']
merged_df['Motor Vehicle Theft']=merged_df['PercentChange_Motor_vehicle_theft']
merged_df['Arson']=merged_df['PercentChange_Arson']
```

```python
#Chart largest percent change by violent crime types
def update_chart(selected_column):
    # Sort the DataFrame based on the selected column and select the top 10 rows
    top_10_df = merged_df.sort_values(by=selected_column, ascending=False).head(10)
    
    plt.figure(figsize=(10, 6))
    plt.bar(top_10_df['City'], top_10_df[selected_column])
    plt.xlabel('City')
    plt.ylabel('Percent Change')
    plt.title(f'Largest Percent Change in {selected_column} by City')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Create the interactive dropdown box
column_dropdown = widgets.Dropdown(
    options=['Violent Crime','Murder', 'Rape','Robbery','Aggravated Assault','Property Crime', 'Burglary','Larceny-Theft','Motor Vehicle Theft','Arson'],
    value='Violent Crime',
    description='Select Column:'
)

# Connect the dropdown to the update_chart function
interact(update_chart, selected_column=column_dropdown)
plt.show()
```
